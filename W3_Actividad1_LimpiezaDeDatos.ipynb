# Parte 1 - Fundamentos de Bases de Datos

### Fundamentos de bases de datos y para ciencia de datos.  
### Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos

Un warehouse es una base de datos que contiene información de diferentes fuentes y las une. La información en el warehouse puede fluir de una o varios origines.
Para acceder a los datos existen dos paradigmas principales que son ETL y ELT. Por sus siglas en inglés son los mismos términos; Extraer, Leer  y transformar (Extract, Load, Transform). ETL; Extraer, Leer y Transformar, es la manera convensional de consumir los datos. Éste paradigma se utiliza hasta que la cantidad de datos se vuelve un problema de escalabilidad. Es entonces que utilizamos ELT; extraer, leer y transformar. Ésta manera de consumir datos es conveniente cuando tenemos largos volúmenes de información no estructurada. De modo que el contenido puede utilizarse de diferentes maneras a como sea conveniente.
Términos comunes en fundamentos de bases de datos también son los tipos de datos; estructurados y no estructurados. Los primeros se refieren a tablas bien comprimidas de información donde cada fila o registro representa una colección de información o evento. Esto puede extenderse según las columnas, campos o atributos que el evento contenga. A toda esta información se le conoce como tabla o entidad. Los datos no estructurados se refieren a imágenes, texto, audio o información que se almacena de manera crudo o no procesada.

#### Primer método de limpieza - Eliminación por lista
Se eliminaron los datos NAN de las primeras columnas porque la cantidad de datos es mínima y representa eliminar el posible sesgo al sustituir la información. 5 registros removidos. 

```python
df.dropna(subset=['X2', 'X3', 'X4', 'X5', 'Y'], inplace=True)
df
```

#### Segundo método de limpieza

Sustitución mediana

```python
columnsToSubstitute = ['X6', 'X7', 'X8', 'X9', 'X10', 'X11', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X19', 'X20', 'X21', 'X22', 'X23']
for col in columnsToSubstitute:
  mediana = df[col].median()
  df[col] = df[col].fillna(mediana)
```

```python
df.isnull().sum()
```
